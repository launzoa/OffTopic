```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def plot_2D(t, x, y, label):                    
    plt.figure()
    plt.plot(x, y, label=label, color='blue')
    plt.title('Gráfico de funções')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axvline(0, color='gray', linewidth=0.8) 
    plt.axhline(0, color='gray', linewidth=0.8) 
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.axis('equal')
    plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
t = np.linspace(-5, 5, 100)
```

```python
x = np.ones_like(t)
y = t
plot_2D(t, x, y, label=r'$F(t) = (1, t)$')
```

```python
x = 2*t - 1
y = t + 2
plot_2D(t, x, y, r"$F(t) = (2t − 1, t + 2)$")
```

```python
x = t**2 
y = t 
plot_2D(t, x, y, r"$F(t) = (t², t)$")
```

```python
x = np.cos(t)
y = 2 * np.sin(t)
plot_2D(t, x, y, r"F = (cos(t), 2sin(t))") 
```

```python
x = np.sin(t)
y = np.sin(t)**2
plot_2D(t,x,y,r"$F(t) = (sin(t), sin²(t))$")
```

```python
t = np.linspace(0,10, 100)
x = np.exp(t) * np.cos(t)
y = np.exp(t) * np.sin(t)
print(np.exp(np.pi) * np.cos(np.pi))
print(np.exp(np.pi/2) * np.sin(np.pi/2))
plot_2D(t,x,y,r"$F(t) = (e^tcos(t), e^tsin(t))$")
```
