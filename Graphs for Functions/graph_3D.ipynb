```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

plt.style.use('seaborn-v0_8-darkgrid')

def plot_3d_curve(t_values, x_values, y_values, z_values,
                  label_plot="", linewidth=2,
                  xlabel='X', ylabel='Y', zlabel='Z'):

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.plot(x_values, y_values, z_values, label=label_plot, linewidth=linewidth)

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_zlabel(zlabel)

    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout() 
    plt.show()
```

```python
# a) F(t) = (1, t, 1) 
t = np.linspace(-5, 5, 200)
x = np.ones_like(t)
y = t
z = np.ones_like(t)
plot_3d_curve(t, x, y, z,label_plot=r'$F(t) = (1, t, 1)$')
```

```python
# c) F(t) = (t, t, 1), t ≥ 0
t = np.linspace(0, 5, 200) 
x = t
y = t
z = np.ones_like(t)
plot_3d_curve(t, x, y, z, label_plot=r'$F(t) = (t, t, 1)$')
```

```python
# e) F(t) = (t, t, 1 + sen t), t ≥ 0
t = np.linspace(0, 6 * np.pi, 500) 
x = t
y = t
z = 1 + np.sin(t)
plot_3d_curve(t, x, y, z,label_plot=r'$F(t) = (t, t, 1 + \sin t)$') 
```

```python
# g) F(t) = (cos t, sen t, 2)
t = np.linspace(0, 2 * np.pi, 200)
x = np.cos(t)
y = np.sin(t)
z = np.full_like(t, 2) 
plot_3d_curve(t, x, y, z, label_plot=r'$F(t) = (\cos t, \sin t, 2)$') 
```

```python
# i) F(t) = (t, t, 1/t), t > 0
t = np.linspace(0.1, 5, 200) 
x = t
y = t
z = 1 / t
plot_3d_curve(t, x, y, z,label_plot=r'$F(t) = (t, t, 1/t)$') 
```

```python
# l) F(t) = (e^(-t) cos t, e^(-t) sen t, e^(-t)), t ≥ 0
t = np.linspace(0, 6 * np.pi, 500) 
x = np.exp(-t) * np.cos(t)
y = np.exp(-t) * np.sin(t)
z = np.exp(-t)
plot_3d_curve(t, x, y, z,label_plot=r'$F(t) = (e^{-t} \cos t, e^{-t} \sin t, e^{-t})$') 
```

```python
# n) F(t) = (sen t, sen t, t), t ≥ 0
t = np.linspace(0, 8 * np.pi, 500) 
x = np.sin(t)
y = np.sin(t)
z = t
plot_3d_curve(t, x, y, z,label_plot=r'$F(t) = (\sin t, \sin t, t)$') 
```
